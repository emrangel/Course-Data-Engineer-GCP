```python
from datetime import datetime
import uuid
from faker import Faker
import random

fake = Faker()

for i in range(25):
    cantidad = random.randint(1,30)
    precioUni = round(random.uniform(10, 1000), 2)
    payload = {
            "IdTrx" : str(uuid.uuid4()),
            "Producto" : fake.vin(), 
            "Cantidad" : cantidad, 
            "PrecioUni" : precioUni,
            "PrecioTot" : round(cantidad * precioUni,2),
            "Cliente" : fake.ascii_free_email(),
            "Direccion" : fake.street_address(),
            "HoraTrx" : datetime.now().strftime('%Y-%m-%d %H:%M:%S')        
        } 
   
    coleccion_usuarios = db.collection("transacciones")
    coleccion_usuarios.add(payload)

    print(f"Documento insertado con exito:\n {payload}")
```

```python
from google.cloud import firestore

# Crea una instancia del cliente Firestore
db = firestore.Client(database='ronny-dev')

# Referencia a la colección que deseas consultar
coleccion = db.collection("transacciones")

# Realiza una consulta simple (por ejemplo, filtrando documentos)
query = coleccion.where("Cliente", "==", "martinezkelly@hotmail.com")

# Ejecuta la consulta y obtén los resultados
resultados = query.stream()

# Itera a través de los resultados
for resultado in resultados:
    print(f"ID del documento: {resultado.id}, Datos: {resultado.to_dict()}")
```
