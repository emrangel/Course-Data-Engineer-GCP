```python
columns = ['order_id', 'order_date', 'order_customer_id', 'order_status']
```

Subimos el archivo orders.csv o cualquier archivo csv que tengamos a la mano.

```python
df = pd.read_csv('orders.csv', names=columns)
df
```

```python
df.to_csv(
    f"gs://{bucket_name}/orders.csv",
    index=False
)
```

```python
df.to_parquet(
    f"gs://{bucket_name}/orders.parquet",
    index=False
)
```

```python
df_parquet = pd.read_parquet(f"gs://{bucket_name}/orders.parquet")
df_parquet
```
