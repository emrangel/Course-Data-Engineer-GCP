# Google Cloud Storage - Python API

La biblioteca ***google-cloud-storage*** es una librería de Python para interactuar con Google Cloud Storage, un servicio de almacenamiento de objetos en la nube de Google. Proporciona una forma eficiente y fácil de manejar datos a gran escala. Aquí te presento algunas de las funciones más utilizadas y sus explicaciones:

**1.Inicialización del Cliente**

Primero, necesitas inicializar un cliente para interactuar con el servicio de Google Cloud Storage.

```python
from google.cloud import storage

client = storage.Client(project='project-d0c76d9f-d27a-4298-8e3')
```

**2. Crear un Bucket:**

Los 'buckets' son contenedores básicos donde se almacenan tus datos en Cloud Storage.

```python
bucket_name = 'sd-consola-bucket-emra-02'
bucket = client.create_bucket(bucket_name)
```

**3. Subir Archivos (Blobs)**

Los objetos almacenados en Google Cloud Storage se conocen como blobs. Para subir un archivo, primero necesitas un blob.

```python
!ls ./sample_data
```

```python
bucket = client.get_bucket(bucket_name)
blob = bucket.blob('prefix-01/orders.csv')
blob.upload_from_filename('orders.csv')
```

**4. Descargar Archivos**

Puedes descargar archivos (blobs) a tu sistema local.

```python
blob = bucket.blob('prefix-01/orders.csv')
blob.download_to_filename('orders_downloaded.csv')
```

**5. Listar Buckets**

Puedes listar todos los buckets disponibles en tu proyecto.

```python
buckets = list(client.list_buckets())
buckets
```

**6. Listar Archivos en un Bucket**

Obtén una lista de todos los blobs en un bucket específico.

```python
blobs = bucket.list_blobs()
for blob in blobs:
    print(blob.name)
```

**7. Eliminar Archivos y Buckets**

Puedes eliminar blobs (archivos) y buckets completos.

Eliminar un archivo:

```python
blob = bucket.blob('prefix-01/orders.csv')
blob.delete()
```

Eliminar un bucket:

```python
bucket.delete()
```
